```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

qc = QuantumCircuit(2, 2)
qc.h(0)        
qc.cx(0, 1)    
qc.measure([0,1], [0,1])
display(qc.draw(output='mpl'))

simulator = AerSimulator()
result = simulator.run(qc).result()
counts = result.get_counts()
print("Ideal Simulation Results:",counts)
plot_histogram(counts)

fig1_circuit = qc.draw(output='mpl')
fig1_circuit.savefig('quantum_circuit.png', dpi=300, bbox_inches='tight')

ideal_counts = result.get_counts()

fig1_ideal_res = plot_histogram(ideal_counts)
fig1_ideal_res.savefig('results_histogram.png', dpi=300, bbox_inches='tight')
```

```python

from qiskit_aer.noise import NoiseModel
from qiskit_ibm_runtime.fake_provider import FakeManilaV2

device_backend = FakeManilaV2()
noise_model = NoiseModel.from_backend(device_backend)

simulator = AerSimulator(noise_model=noise_model)
result_noisy = simulator.run(qc, shots=1024).result()
counts_noisy = result_noisy.get_counts()

print("Noisy Simulation Results (IBM Manila):", counts_noisy)
plot_histogram(counts_noisy)


fig1_noisy_res = plot_histogram(counts_noisy)
fig1_noisy_res.savefig('noisy_results_histogram.png', dpi=300, bbox_inches='tight')
```
