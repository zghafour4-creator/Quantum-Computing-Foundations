```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

dj_oracle = QuantumCircuit(2)
dj_oracle.cx(0, 1) 

dj_circuit = QuantumCircuit(2, 1)

dj_circuit.x(1)
dj_circuit.h(1)

dj_circuit.h(0)
dj_circuit.barrier()

dj_circuit.compose(dj_oracle, inplace=True)
dj_circuit.barrier()

dj_circuit.h(0)

dj_circuit.measure(0, 0)

display(dj_circuit.draw(output='mpl'))

simulator = AerSimulator()
counts = simulator.run(dj_circuit).result().get_counts()

print("Measurement results (Balanced Oracle):", counts)
plot_histogram(counts)

fig3_circuit = dj_circuit.draw(output='mpl')
fig3_circuit.savefig('deutsch_jozsa_circuit.png', dpi=300, bbox_inches='tight')

fig3_res = plot_histogram(counts)
fig3_res.savefig('deutsch_jozsa_results.png', dpi=300, bbox_inches='tight')
```
