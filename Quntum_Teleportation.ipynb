```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

qc = QuantumCircuit(3, 3)

qc.x(0)
qc.barrier()

qc.h(1)
qc.cx(1, 2)
qc.barrier()

qc.cx(0, 1)
qc.h(0)
qc.barrier()

qc.measure([0, 1], [0, 1])
qc.barrier()

qc.cx(1, 2)
qc.cz(0, 2)

qc.measure(2, 2)

display(qc.draw(output='mpl'))

simulator = AerSimulator()
counts = simulator.run(qc).result().get_counts()
plot_histogram(counts)

fig2_circuit = qc.draw(output='mpl')
fig2_circuit.savefig('teleportation_circuit.png', dpi=300, bbox_inches='tight')

fig2_res = plot_histogram(counts)
fig2_res.savefig('teleportation_results.png', dpi=300, bbox_inches='tight')
```
